## Aggregate NDT Metrics for USA over Q3 2012

```python
import pandas
import MySQLdb
import pandas.io.sql as psql
import matplotlib
from datetime import datetime as dt
from bq2ipy import bigquery
import socket
import os

START_DATE="2012-07-01"
END_DATE="2012-12-31"

def get_short_host_and_ip_map(filename):
    f = open(filename, 'r')
    hmap = {}
    for line in f.readlines():
        (h,ip) = line.split()
        h = ".".join(h.split(".")[2:4])
        hmap[h] = ip
    return hmap

def get_bq_table_names_from_dates(start, end):
    months=[] # ['m_lab.2012_07', 'm_lab.2012_08', 'm_lab.2012_09', 'm_lab.2012_10' ]
    beginning = datetime.datetime.strptime(start, "%Y-%m-%d") 
    ending    = datetime.datetime.strptime(end, "%Y-%m-%d") 
    years = range(beginning.year, ending.year+1)
    month_curr = beginning.month
    for year_curr in years:
        from_date = datetime.datetime(year_curr, month_curr, 1)
        end_of_year = datetime.datetime(year_curr, 12, 31)
        if ending < end_of_year:
            to_date = ending
        else:
            to_date = end_of_year
        months += [ "m_lab.%s_%02d" % (year_curr, m) for m in range(from_date.month, to_date.month+1)]
    return months

host_map   = get_short_host_and_ip_map("server-ips-usa.txt")
month_list = get_bq_table_names_from_dates(START_DATE, END_DATE)
```

## Queries

```python
VALID_DOWNLOAD_TEST="""
        IS_EXPLICITLY_DEFINED(project)
    AND project = 0
    AND IS_EXPLICITLY_DEFINED(connection_spec.data_direction)
    AND connection_spec.data_direction = 1
    AND IS_EXPLICITLY_DEFINED(web100_log_entry.is_last_entry)
    AND web100_log_entry.is_last_entry = True
    AND IS_EXPLICITLY_DEFINED(web100_log_entry.snap.HCThruOctetsAcked)
    AND web100_log_entry.snap.HCThruOctetsAcked >= 8192
    AND web100_log_entry.snap.HCThruOctetsAcked < 1000000000
    AND (web100_log_entry.snap.SndLimTimeRwin +
         web100_log_entry.snap.SndLimTimeCwnd +
         web100_log_entry.snap.SndLimTimeSnd) >= 9000000
    AND (web100_log_entry.snap.SndLimTimeRwin +
         web100_log_entry.snap.SndLimTimeCwnd +
         web100_log_entry.snap.SndLimTimeSnd) < 3600000000
    AND IS_EXPLICITLY_DEFINED(web100_log_entry.snap.CongSignals)
    AND web100_log_entry.snap.CongSignals > 0
    AND IS_EXPLICITLY_DEFINED(web100_log_entry.connection_spec.remote_ip)
    AND IS_EXPLICITLY_DEFINED(web100_log_entry.connection_spec.local_ip)
"""

def query_ndt_downloads(iplist, months):
    args = {}
    IPS=",".join(["'%s'" % ip for ip in iplist ]) # create a format like: 'ip1','ip2','ip3',...
    MONTHS=",".join(months)
    
    # direction = 1 is download.
    query="""
select  
    UTC_USEC_TO_MONTH(web100_log_entry.log_time*1000000) as date,
    count(test_id) as test_count,
    web100_log_entry.connection_spec.local_ip as server_ip,
    count(distinct web100_log_entry.connection_spec.remote_ip) as unique_clients,
    count(web100_log_entry.connection_spec.remote_ip) as total_clients, 
    SUM(web100_log_entry.snap.HCThruOctetsAcked) as total_bytes,
    SUM(web100_log_entry.snap.SndLimTimeRwin +
        web100_log_entry.snap.SndLimTimeCwnd +
        web100_log_entry.snap.SndLimTimeSnd) as total_time
from    
    %(months)s
where 
    %(valid_download_test)s
    AND web100_log_entry.connection_spec.local_ip IN(%(ip_list)s) 
group by
    date, server_ip
order by
    date; 
""" % { 'ip_list' : IPS, 
        'months' : MONTHS, 
        'valid_download_test' : VALID_DOWNLOAD_TEST }

    #print query
    ret = bigquery(query)
    print "query returned ok"
    return ret

def query_ndt_unique_ip_frequency(iplist, months):
    args = {}
    IPS=",".join(["'%s'" % ip for ip in iplist ]) # create a format like: 'ip1','ip2','ip3',...
    MONTHS=",".join(months)
    seconds_per_week = 604800  # bin dates into 7 day blocks
    query="""
select  
    UTC_USEC_TO_MONTH(web100_log_entry.log_time*1000000) as date,
    web100_log_entry.connection_spec.remote_ip as unique_ip,
    count(test_id) as test_count,
    SUM(web100_log_entry.snap.HCThruOctetsAcked) as total_bytes,
    SUM(web100_log_entry.snap.SndLimTimeRwin +
        web100_log_entry.snap.SndLimTimeCwnd +
        web100_log_entry.snap.SndLimTimeSnd) as total_time
from    
    %(months)s
where 
    %(valid_download_test)s
    AND web100_log_entry.connection_spec.local_ip IN(%(ip_list)s) 
group by
    date, unique_ip
order by
    date; 
""" % { 'ip_list' : IPS, 'months' : MONTHS, 
        'sec_per_week' : seconds_per_week,  
        'valid_download_test' : VALID_DOWNLOAD_TEST}
    #print query
    ret = bigquery(query)
    print "query returned ok"
    return ret


def query_ndt_unique_ip_max_average_with_min(iplist, months, minimum_repeated_tests):
    args = {}
    IPS=",".join(["'%s'" % ip for ip in iplist ]) # create a format like: 'ip1','ip2','ip3',...
    MONTHS=",".join(months)
    seconds_per_week = 604800  # bin dates into 7 day blocks
    query="""
select  
    UTC_USEC_TO_MONTH(web100_log_entry.log_time*1000000) as date,
    web100_log_entry.connection_spec.remote_ip as unique_ip,
    count(test_id) as test_count,
    max(web100_log_entry.snap.HCThruOctetsAcked/(web100_log_entry.snap.SndLimTimeRwin +
        web100_log_entry.snap.SndLimTimeCwnd +
        web100_log_entry.snap.SndLimTimeSnd)) as max_rate,
    SUM(web100_log_entry.snap.HCThruOctetsAcked) as total_bytes,
    SUM(web100_log_entry.snap.SndLimTimeRwin +
        web100_log_entry.snap.SndLimTimeCwnd +
        web100_log_entry.snap.SndLimTimeSnd) as total_time
from    
    %(months)s
where 
    %(valid_download_test)s
    AND web100_log_entry.connection_spec.local_ip IN(%(ip_list)s) 
group by
    date, unique_ip
having
    test_count >= %(minimum_repeated_tests)d
order by
    date;
""" % { 'ip_list' : IPS, 'months' : MONTHS, 
        'sec_per_week' : seconds_per_week,  
        'valid_download_test' : VALID_DOWNLOAD_TEST,
        'minimum_repeated_tests' : minimum_repeated_tests}
    #print query
    ret = bigquery(query)
    print "query returned ok"
    return ret


def query_avg_ndt_downloads(iplist, months):
    args = {}
    IPS=",".join(["'%s'" % ip for ip in iplist ]) # create a format like: 'ip1','ip2','ip3',...
    MONTHS=",".join(months)
    
    # direction = 1 is download.
    query="""
select  
    UTC_USEC_TO_MONTH(web100_log_entry.log_time*1000000) as date,
    count(test_id) as test_count,
    count(distinct web100_log_entry.connection_spec.remote_ip) as unique_clients,
    count(web100_log_entry.connection_spec.remote_ip) as total_clients, 
    SUM(web100_log_entry.snap.HCThruOctetsAcked) as total_bytes,
    SUM(web100_log_entry.snap.SndLimTimeRwin +
        web100_log_entry.snap.SndLimTimeCwnd +
        web100_log_entry.snap.SndLimTimeSnd) as total_time
from    
    %(months)s
where 
    %(valid_download_test)s
    AND web100_log_entry.connection_spec.local_ip IN(%(ip_list)s) 
group by
    date
order by
    date; 
""" % { 'ip_list' : IPS, 
        'months' : MONTHS, 
        'valid_download_test' : VALID_DOWNLOAD_TEST }

    #print query
    ret = bigquery(query)
    print "query returned ok"
    return ret
```

## Calculate Peak-Average and Average-Average

```python
NAME=host_map.keys()
IPS=host_map.values()

host_map   = get_short_host_and_ip_map("server-ips-usa.txt")
month_list = get_bq_table_names_from_dates(START_DATE, END_DATE)
rumap={}
for m in month_list:
    rumap[m] = query_ndt_unique_ip_max_average_with_min(host_map.values(), [m], 5)
```

```python
peak_total = 0.0
peak_count = 0.0
average_total = 0.0
average_count = 0.0

for m in month_list:
    if m in rumap:
        for i in range(len(rumap[m]['max_rate'])):
            peak_total += 8.0*rumap[m]['max_rate'][i]
            peak_count += 1
            average_total += 8.0*(rumap[m]['total_bytes'][i] / rumap[m]['total_time'][i])
            average_count += 1 
    
    print "average peak:" , peak_total/peak_count
    print "average:     " , average_total/average_count
    print "ratio:       " , ((average_total/average_count)/(peak_total/peak_count))*100.0
```
